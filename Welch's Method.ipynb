```python
import numpy as np
from scipy.signal import welch
import matplotlib.pyplot as plt
import pandas
```

## Import Data

```python
power_time_series = pandas.read_csv("test.csv")["power"]
plt.plot(power_time_series)
plt.xlabel("sample")
plt.ylabel("power (W)")
```

```python
# Assume power_time_series is your measured power over time
fs = 500/13 # Sampling frequency (inverse of your time step)

f, Pxx = welch(power_time_series, fs=fs, nperseg=1024)  # Welch PSD

plt.plot(f, Pxx)  # Log-log plot for better visualization
plt.xlabel('Frequency [Hz]')
plt.ylabel('PSD [arb. units]')
plt.title('Power Spectral Density of Laser Drift')
plt.show()
```
